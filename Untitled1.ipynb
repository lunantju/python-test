```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
img_cat=cv2.imread('WechatIMG8621.jpeg')
img_cat
```

```python
#题目一
b,g,r=cv2.split(img_cat)

r1=(r-np.mean(r))/np.std(r,ddof=0)
g1=(r-np.mean(g))/np.std(g,ddof=0)
b1=(r-np.mean(b))/np.std(b,ddof=0)

img1=cv2.merge((b1,g1,r1))
img1

```
